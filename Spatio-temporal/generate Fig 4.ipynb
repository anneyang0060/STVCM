```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid", palette="pastel", color_codes=True)
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
from itertools import product
import multiprocessing as mp
import os
```

```python
file = 'V1_CityF_pool.csv' # 'V1_CityE_expand_withID.csv'  
df = pd.read_csv('C:/Users/annie/Spatio-temporal/data/'+file, index_col=['date','time'])
```

```python
grids = df['grid_id'].unique()
G = len(grids)
```

```python
for g in range(G):
    df_sub = df[df['grid_id'] == grids[g]]
    df_sub = df_sub.fillna(0)
    
    fig = plt.figure(figsize=(6, 4))
    df_sub.groupby('date')['gmv'].plot(use_index=False)
    plt.title('Drivers\' Total Income')
```
