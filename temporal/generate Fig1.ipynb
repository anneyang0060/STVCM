```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid", palette="pastel", color_codes=True)
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
from itertools import product
import multiprocessing as mp
import os
```

```python
file = 'V2_cityA_serial_order_dispatch_AA.csv' 
df1 = pd.read_csv('C:/Users/annie/temporal/data/'+file, index_col=['date','time'])
df1 = df1.groupby(['date','time']).sum()
```

```python
file = 'V2_cityB_serial_order_dispatch_AA.csv'
df2 = pd.read_csv('C:/Users/annie/temporal/data/'+file, index_col=['date','time'])
df2 = df2.groupby(['date','time']).sum()
```

```python
fig = plt.figure(figsize=(14, 7))

plt.subplot(2,3,1)
df1.groupby('date')['gmv'].plot(use_index=False)
plt.title('Drivers\' Total Income')

plt.subplot(2,3,2)
df1.groupby('date')['cnt_call'].plot(use_index=False)
plt.title('Number of Requests')

plt.subplot(2,3,3)
df1.groupby('date')['sum_online_time'].plot(use_index=False)
plt.title('Drivers\' Total Online Time')

plt.subplot(2,3,4)
df2.groupby('date')['gmv'].plot(use_index=False)
plt.title('Drivers\' Total Income')

plt.subplot(2,3,5)
df2.groupby('date')['cnt_call'].plot(use_index=False)
plt.title('Number of Requests')

plt.subplot(2,3,6)
df2.groupby('date')['sum_online_time'].plot(use_index=False)
plt.title('Drivers\' Total Online Time')

name = 'C:/Users/annie/temporal/fig/cityAB_t_pattern.pdf'
plt.savefig(name, dpi=200, bbox_inches='tight')
```
